```python
# 1. Swap values of two variables without using a third variable
a = int(input("Enter first number (a): "))
b = int(input("Enter second number (b): "))
a, b = b, a
print("After swapping: a =", a, ", b =", b)

# 2. Take name, age, and GPA as input and print them formatted
name = input("Enter your name: ")
age = int(input("Enter your age: "))
gpa = float(input("Enter your GPA: "))
print(f"Name: {name}, Age: {age}, GPA: {gpa:.2f}")

# 3. Take two numbers and print sum, difference, product, and division
x = float(input("Enter first number: "))
y = float(input("Enter second number: "))
print("Sum =", x + y)
print("Difference =", x - y)
print("Product =", x * y)
print("Division =", x / y)

# 4. Take age and check if eligible for voting (age â‰¥ 18)
age = int(input("Enter your age: "))
if age >= 18:
    print("You are eligible for voting.")
else:
    print("You are not eligible for voting.")

# 5. Check if '#' exists in "apple"
if "#" in "apple":
    print("Yes, '#' exists in 'apple'")
else:
    print("No, '#' does not exist in 'apple'")

# 6. Check if a number is even or odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")

# 7. Take marks as input and print grade
marks = float(input("Enter your marks: "))
if marks >= 80:
    print("Grade: A")
elif marks >= 60:
    print("Grade: B")
elif marks >= 40:
    print("Grade: C")
else:
    print("Grade: Fail")

# 8. Print numbers from 1 to 20 using a for loop
print("Numbers from 1 to 20:")
for i in range(1, 21):
    print(i, end=" ")
print()

# 9. Print multiplication table of any number using a while loop
n = int(input("\nEnter a number for multiplication table: "))
i = 1
while i <= 10:
    print(f"{n} x {i} = {n*i}")
    i += 1

# 10. Function to check if a number is prime or not
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

n = int(input("Enter a number to check if it's prime: "))
print("Prime" if is_prime(n) else "Not Prime")

# 11. Function with default arguments
def greet(name="Guest"):
    print(f"Hello, {name}! Welcome.")

greet()          # default
greet("Salar")   # with argument

# 12. Lambda function to square a number
square = lambda x: x * x
num = int(input("Enter a number to square: "))
print("Square:", square(num))
```
